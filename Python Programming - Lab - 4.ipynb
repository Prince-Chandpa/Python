### 06) WAP to print sum of series 1 + 4 + 9 + 16 + 25 + 36 + ...n.

```python
n = int(input('Enter Number:'))

sum = 0

for i in range(1, n+1):
  if i == n:
    print(i*i,end=' = ')
  else:
    print(i*i,end=' + ')
  sum += i*i

print(sum)
```

### 07) WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
n = int(input('Enter Number:'))
sum = 0

for i in range(1, n+1):
  if i%2 == 0:
    if i == n:
      print(i,end=' = ')
    else:
      print(i,end=' + ')
    sum -= i
  else:
    if i == n:
      print(i,end=' = ')
    else:
      print(i,end=' - ')
    sum += i
print(sum)
```

### 10) WAP to print GCD of given two numbers.

```python
a = int(input('Enter first Number:'))
b = int(input('Enter second Number:'))

min = a if a<b else b

gcd = 1

for i in range(1,min+1):
  if (a%i == 0 and b%i == 0):
    gcd = i
print(f'GCD({a},{b}) = {gcd}')
```

### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input('Enter Number: '))
is_prime = True
for i in range(2,n-1):
  if(n%i == 0):
    is_prime = False
    break

if is_prime:
  print('Prime')
else:
  print('Not Prime')
```

### 13) WAP to print sum of digits of given number.

```python
n = int(input('Enter Number: '))
sum = 0
while n > 0:
  sum += n%10
  n //= 10
print(f'Sum = {sum}')
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input('Enter Number: '))
num = n
rev = 0
while n > 0:
  rev = rev*10 + n%10
  n //= 10

print('Palindrome' if rev == num else 'Not Palindrome')
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n = int(input('Enter Number: '))
num = n
digit = 0
sum = 0

while n > 0:
  digit += 1
  n //= 10

n = num

while n > 0:
  sum = sum + ((n%10)**digit)
  n //= 10

print('Armstrong' if sum == num else 'Not Armstrong')
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
num = int(input('Enter Number: '))

for n in range(1,num+1):
  sum = 0
  for i in range(1,n):
    if n%i == 0:
      sum += i
  if sum == n:
    print(n)
```
