### 01) WAP to find sum of all the elements in a List.

```python
l = list(map(int,input('Enter element: ').split()))
sum = 0
for i in l:
    sum += i

print(f'SUM: {sum}')
```

### 02) WAP to find largest element in a List.

```python
l = input('Enter element: ').split()
l = [int(i) for i in l]
large = 0

for i in l:
    if i > large:
        large = i

print('Large:',large)
```

### 09) WAP to count unique items in a list.

```python
l = list(map(int,input('Enter element: ').split()))


count_unique = 0

for i in l:
    count = 0    
    for j in l:
        if i == j:
            count += 1
    if count <=1:
        count_unique += 1
print(count_unique)
```

### 12) WAP to count occurrences of an element in a list.

### 13) WAP to find second largest number in a list.

```python
l = list(map(int,input('Enter element: ').split()))
# l.sort()
first_largest = l[0]
second_largest = l[0]

for i in l:
    if i > first_largest :
        second_largest = first_largest
        first_largest = i
    elif i > second_largest and i != first_largest:
        second_largest = i
print(f'First Largest: {first_largest}, Second Largest: {second_largest}')
```

### 14) WAP to extract elements with frequency greater than K.

```python
l = list(map(int,input('Enter element: ').split()))
k = int(input('Enter K: '))

count_list = []
for i in l:
    frequency = 0    
    for j in l:
        if i == j:
            frequency += 1
    if frequency > k and i not in count_list:
        count_list.append(i)
print(count_list)
```

### 15) WAP to create a list of squared numbers from 0 to 9 with and without using List Comprehension.

```python
without_l = []

with_l = [i**2 for i in range(10)]

for i in range(10):
    without_l.append(i**2)

print('Without:',without_l)
print('With:',with_l)
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

### 17) WAP to create a list of common elements from given two lists.

```python
l1 = list(map(int,input('Enter element of list 1: ').split()))
l2 = list(map(int,input('Enter element of list 2: ').split()))
ans_list = []
for i in l1:
    for j in l2:
        if i == j and i not in ans_list:
            ans_list.append(i)
print(ans_list)
```
