### 04) WAP to accept name (string), age (int), and percentage (float).

### 15) WAP to get the distance from user into kilometer, and convert it into meter, feet, inches and centimeter.

```python
kilometer = float(input("Enter Distance(in km):"))

meter = kilometer*1000
inch = meter*39.3701
feet = meter*3.28084
centimeter = meter*100


print(f"Kilometer: {kilometer}km")
print(f"meter: {meter}m")
print(f"inch: {inch}\"")
print(f"feet: {feet}\'")
print(f"centimeter: {centimeter}cm")
```
