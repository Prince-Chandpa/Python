### 01) Write a function to calculate BMI given mass and height. (BMI = mass/h**2)

```python
def calculate_bmi(mass,height):
    return (mass/(height**2))

mass = int(input('Enter Mass: '))
height = int(input('Enter Height: '))

bmi = calculate_bmi(mass,height)

print(f'BMI = {bmi}')
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def is_prime(n):
    if n <= 1:
        return 0
    for i in range(2, int(n**0.5) + 1):
        if n%i == 0:
            return 0
    return 1

n = int(input('Enter N: '))
print(f'{n} is {"not " if not is_prime(n) else ""}a Prime Number')
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
def prime_in_range(lower_range,upper_range):
    prime_list = []
    if upper_range < lower_range:
        upper_range,lower_range = lower_range,upper_range
        
    for i in range(lower_range,upper_range):
        if is_prime(i):
            prime_list.append(i)
    return prime_list

lower_range = int(input('Enter Lower Range: '))
upper_range = int(input('Enter Upper Range: '))
print(prime_in_range(lower_range,upper_range))
```

### 07) Write a function to calculate the sum of the first element of each tuples inside the list.

```python
def sum_of_tuple_first(l):
    add = 0
    for t in l:
        add += t[0]
    return add

l = [
    (1,2,3),
    (4,5,6)
]

print(f'sum of tuple first element : {sum_of_tuple_first(l)}')
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
def fibonacci(n,a = 0,b = 1, i = 1):
    if(n == i):
        return a
    return fibonacci(n, b, a+b, i+1)

n = int(input('Enter N: '))
fibonacci(n)
```

### 10) Write a function to get the sum of the scores ending with zero.
#### Example : scores = [200, 456, 300, 100, 234, 678] 
#### Ans = 200 + 300 + 100 = 600

```python
def sum_of_scores(scores):
    add = 0
    for i in scores:
        if i%10 == 0:
            add += i
    return add
    
scores = [int(i) for i in input("Enter list: ").split()]

print(f'Ans = {sum_of_scores(scores)}')
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def reverse_key_value(d):
    return {value:key for key,value in d.items()}

d = {'a': 10, 'b':20, 'c':30, 'd':40}
reverse_dict = reverse_key_value(d)
print(reverse_dict)
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
def count_case(s):
    upper_count = 0
    lower_count = 0

    for i in s:
        if i.isupper():
            upper_count += 1
        elif i.islower():
            lower_count += 1
    return (upper_count,lower_count)

s = input('Enter String: ')

upper_count,lower_count = count_case(s)

print(f'no_upper = {upper_count}, no_lower = {lower_count}')
```
