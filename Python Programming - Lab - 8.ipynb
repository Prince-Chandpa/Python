### 01) WAP to find their union, intersection, and difference of the given two sets.

### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
dict1 = {1:10, 2:20, 3:30}
dict2 = {2:40, 3:50, 4:60}
new_dict = {}
for i in dict1.keys():
    if i in dict2:
        new_dict[i] = dict1[i] + dict2[i]
    else:
        new_dict[i] = dict1[i]

for i in dict2.keys():
    if i not in new_dict:
        new_dict[i] = dict2[i]

print(new_dict)
# key_list = [int(i) for i in input('Enter Keys:').split()]
# value_list = [int(i) for i in input('Enter Value:').split()]
````

### 06) WAP to Sort a dictionary in ascending order of its values.
Hint: in sorted(), use udf for key parameter


**Sample Input:**
```
{'Math': 70, 'Physics': 85, 'Chemistry': 60}

```
**Sample Output:**
```
{'Chemistry': 60, 'Math': 70, 'Physics': 85}
```

````python
key_list = [i for i in input('Enter Keys:').split()]
value_list = [int(i) for i in input('Enter Value:').split()]

dict1 = {i:j for i,j in zip(key_list,value_list)}

def getValues(item):
    return item[1]

dict1 = dict(sorted(dict1.items(),key=getValues))

print(dict1)
````

### 11) WAP to find the student with the highest total marks, given student names and their marks.

**Sample Input:**
```
{'Amit': [70, 80, 90], 'Riya': [85, 75, 80], 'Neha': [60, 70, 65]}
```
**Sample Output:**
```
Topper: Amit
```

````python
marks = {'Amit': [70, 80, 90], 'Riya': [85, 75, 80], 'Neha': [60, 70, 65]}

max_marks = 0
topper = ''

for k,v in marks.items():
    sum_of_mark = sum(v)
    if sum_of_mark > max_marks:
        max_marks = sum_of_mark
        topper = k
print(topper)
````

### 12) WAP to extract keys whose values are even numbers.

**Sample Input:**
```
{1: 11, 2: 20, 3: 15, 4: 40}
```
**Sample Output:**
```
[2, 4]
```

````python
key_list = [int(i) for i in input('Enter Keys:').split()]
value_list = [int(i) for i in input('Enter Value:').split()]

dict1 = {i:j for i,j in zip(key_list,value_list)}

even = [k for k,v in dict1.items() if v%2 == 0]

print(even)
````
