### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
s = input('Enter String: ')
upper_count = 0
lower_count = 0
digit_count = 0
special_count = 0

for i in s:
  if i.isupper():
    upper_count += 1
  elif i.islower():
    lower_count += 1
  elif i.isdigit():
    digit_count += 1
  else:
    special_count += 1

print(f'Uppercase: {upper_count}, Lowercase: {lower_count}, Digits: {digit_count}, Special: {special_count}')
```

### 2) Reverse Each Word
- Input: `Python is Easy` → `nohtyP si ysaE`
- Input: `Hello World` → `olleH dlroW`
- Input: `Learn Python Fast` → `nraeL nohtyP tsaF`

```python
s = input('Enter String: ')
words = s.split()
for i in range(len(words)):
  words[i] = words[i][::-1]
print(' '.join(words))
```

### 8) Replace Vowels

```python
s = input('Enter String: ')
vowels = ['a','e','i','o','u']

for i in s:
  if i.lower() in vowels:
    s = s.replace(i,'*')
print(s)
```

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s = input('Enter String: ')

ans = ''
count = 1

for i in range(1,len(s)):
  if s[i] == s[i-1]:
    count += 1
  else:
    ans += s[i - 1] + str(count)
    count = 1

ans += s[-1] + str(count)

print(ans)
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
s = input('Enter String: ')
ans = ''
for i in s:
  if i.isupper():
    ans += i.lower()
  else:
    ans += i.upper()

print(ans)
```

### 11) Username Validation

1. only alphabets and digits

### 12) Palindrome Checker (ignore the case)
- `Madam` → Palindrome
- `Level` → Palindrome
- `Python` → Not Palindrome

```python
s = input('Enter String: ').lower()

i = 0
j = len(s) - 1

while i < j:
  if s[i] != s[j]:
    print('Not Palindrome')
    break
  i += 1
  j -= 1
else:
  print('Palindrome')
```

### 13) Longest Word Finder
