## Practical 1: Student Marks Analyzer

### Problem Definition
Store student data as a list of tuples. Remove duplicate marks, calculate average marks and store result in a dictionary.

### Sample Input
```python
students = [
    (1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])
]
```

### Sample Output
```
{1: ('Amit', 84.33), 2: ('Neha', 86.33), 3: ('Ravi', 75.0)}

```

````python
students = [(1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])]

students_dict = {}

for k,name,mark in students:
    mark = set(mark)
    avg_of_mark = round(sum(mark)/len(mark),2)
    students_dict[k] = name,avg_of_mark

# print(students)
print(students_dict)
````

## Practical 2: Inventory Management

### Problem Definition
Remove out-of-stock items and calculate total inventory value.

### Sample Input
```python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}
```

### Sample Output
```
{'pen': (10, 50), 'pencil': (5, 100)}
Total Value: 1000
```

````python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}
new_inventory = {}
total_value = 0
for name,value in inventory.items():
    if value[1] != 0:
        new_inventory[name] = value
        total_value += value[0]*value[1]
print(new_inventory)
print(f'Total Value: {total_value}')
````

## Practical 3: Email Domain Extractor

### Problem Definition
Extract email domains and count occurrences.

### Sample Input
```python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
```

### Sample Output
```
{'gmail.com': 2, 'yahoo.com': 1}
```

````python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
domains = [domain.split('@')[1] for domain in emails]
email_domain = {domain:domains.count(domain) for domain in domains}

print(email_domain)
````

## Practical 4: Shopping Cart Billing

### Problem Definition
Calculate item-wise cost and total bill.

### Sample Input
```
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}
```

### Sample Output
```
{'apple': 100, 'banana': 30}
Total Bill: 130
```

````python
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}

bill_dict = {cart_item:cart_count*prices[cart_item] for cart_item,cart_count in cart}

total_bill = sum([total_price for total_price in bill_dict.values()])

print(bill_dict)
print(f'Total Bill: {total_bill}')
````

## Practical 6: Survey Data Aggregation

### Problem Definition
Count responses city-wise and choice-wise.

### Sample Input
```python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]
```

### Sample Output
```
{'Rajkot': {'Yes': 1, 'No': 1}, 'Ahmedabad': {'Yes': 1}}
```

````python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]

city_responses_dict = {}

for name, city, response in responses:
    if city not in city_responses_dict:
        city_responses_dict[city] = {}
    if response not in city_responses_dict[city]:
        city_responses_dict[city][response] = 0
    city_responses_dict[city][response] += 1

print(city_responses_dict)
````
