### 08) WAP to create an UDF named frequency to count occurances of the specific word in a given text file.

```python
def frequency(file_name,word):
    with open(file_name,"r") as fp:
        count_fre = fp.read().count(word)
        fp.close()
    return count_fre

frequency("hello.txt","Hello")
```

### 09) WAP to get the score of five subjects from the user, store them in a file. Fetch those marks and find the highest score.

```python
scores = [i for i in input("Enter you score:").split()]
with open("marks.txt","w") as fp:
    fp.write("\n".join(scores))
    fp.close()

with open("marks.txt","r") as fp:
    max_score = 0
    for i in fp.read().split():
        if int(i) > max_score:
            max_score = int(i)
    print(max_score)
    fp.close()
```

### 10) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2,int(n**0.5)):
        if n % i == 0:
            return False
    return True
```

```python
def first_prime(file_name):
    count_it = 0
    n = 1
    with open(file_name,"w") as fp:
        while count_it != 100:
            if is_prime(n):
                fp.write(str(n)+"\n")
                count_it += 1
            n += 1
        fp.close()
first_prime("primenumbers.txt")
```

### 12) WAP to replace word1 by word2 of a text file. Write the updated data to new file.
