### Practical 1: Count Lines, Words and Characters

**Problem Definition:**  
Write a Python program to read text from a file and count the total number of lines, words, and characters. The result should be written to an output file.


```python
with open('new.txt','r') as fp:
    no_of_line = len(fp.readlines())
    fp.seek(0)
    no_of_words = len(fp.read().split())
    no_of_chars = fp.tell()
    fp.close()
    
with open('output_1.txt','w') as fp:
    fp.write(f'number of line: {no_of_line}\n')
    fp.write(f'number of words: {no_of_words}\n')
    fp.write(f'number of characters: {no_of_chars}')
    print('completed')
    fp.close()
```


### Practical 2: Search a Word in a File

**Problem Definition:**  
Write a Python program to search a given word in a text file and count its occurrences. Store the result in an output file.


```python
def count_word_in_file(file_name,word):
    count = 0
    with open(file_name) as fp:
        data_file = fp.read()
        count = data_file.count(word)
        fp.close()
    return count

def wirte_output_in_file(file_name,word):
    count = count_word_in_file(file_name,word)
    with open('output_1.txt','a') as fp:
        if count == 0:
            fp.write(f'\n{word} not found')
        else:
            fp.write(f'\ncount of {word} = {count}')
    print('completed')
        

wirte_output_in_file('new.txt','this')
```


### Practical 3: Student Marks and Average

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Store marks in a list, calculate the average using a user-defined function, and write student details along with average marks to an output file.


```python
def wirte_file(rol,name,mark):
    with open('student_out.txt','a') as fp:
        mark = mark.split(',')
        mark_int = list(map(int,mark))
        avg = sum(mark_int)/(len(mark_int))
        fp.write(f'{rol} {name} avg: {avg}\n')

with open('students.txt','r') as fp:
    students = fp.read().split('\n')
    for i in students:
        rol,name,mark = tuple(i.split())
        wirte_file(rol,name,mark)
    print('complete')
```


### Practical 4: Maximum and Minimum using Tuple

**Problem Definition:**  
Write a Python program to read numbers from a file, store them in a tuple, and find the maximum and minimum values. Write the result to an output file.


```python
with open('number.txt','r') as fp:
    num_tuple = tuple(map(int,fp.read().split()))
    max_num = max(num_tuple)
    min_num = min(num_tuple)
    fp.close()
with open('output_4.txt','w') as fp:
    fp.write(f'max:{max_num}\nmin:{min_num}')
    print('completed!!')
    fp.close()
```


### Practical 5: Remove Duplicate Words using Set

**Problem Definition:**  
Write a Python program to read words from a file and remove duplicate words using a set. Store the unique words in an output file.


```python
with open('input_5.txt','r') as fp:
    data = fp.read().split()
    data = set(data)
    fp.close()
with open('input_5.txt','w') as fp:
    fp.write(' '.join(data))
    print('complete')
    fp.close()
```


### Practical 6: Word Frequency using Dictionary

**Problem Definition:**  
Write a Python program to count the frequency of each word in a text file using a dictionary and write the result to an output file.


```python
freq = {}
with open('input_5.txt','r') as fp:
    data = fp.read().split()
    for i in data:
        freq[i] = data.count(i)
    fp.close()
with open('output_6.txt','a') as fp:
    for i,j in freq.items():
        fp.write(f'{i} : {j}\n')
    print('complete')
    fp.close()
```


### Practical 7: Student Result and Grade Calculation

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Use a user-defined function to determine grade (A/B/C/Fail) and write the student result to an output file.

**Grade Criteria:**  
- Marks ≥ 75 : Grade A  
- Marks ≥ 60 : Grade B  
- Marks ≥ 40 : Grade C  
- Marks < 40 : Fail


```python
def wirte_file_score(rol,name,mark):
    with open('student_out.txt','a') as fp:
        mark = mark.split(',')
        mark_int = list(map(int,mark))
        avg = sum(mark_int)/(len(mark_int))
        grd = ''
        if avg >= 75:
            grd = 'Grade A'
        elif avg >= 60:
            grd = 'Grade B'
        elif avg >= 40:
            grd = 'Grade C'
        else:
            grd = 'Fail'
        fp.write(f'{rol} {name} avg: {avg} and got {grd}\n')

with open('students.txt','r') as fp:
    students = fp.read().split('\n')
    for i in students:
        rol,name,mark = tuple(i.split())
        wirte_file_score(rol,name,mark)
    print('complete')
```
